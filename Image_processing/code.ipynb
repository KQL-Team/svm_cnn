```python
from PIL import Image
from tqdm import tqdm
from pathlib import Path
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = []
for i in tqdm(range(12567)):
    file_path = f"DATASET/TRAIN/O/O_{i+1}" + '.jpg'
    if Path(file_path).is_file():
        with Image.open(file_path) as image:
            image = image.resize((128,128))
            image = np.array(image)
            image = image.reshape((128, 128, 3))
            print(image.shape if image.shape != (128, 128, 3) else "")
            img.append(image)
```

```python

for i in tqdm(range(9999)):
    file_path = f"DATASET/TRAIN/R/R_{i+1}" + '.jpg'
    if Path(file_path).is_file():
        with Image.open(file_path) as image:
            image = image.resize((128, 128, 3))
            image = np.array(image)
            img.append(image)
```
